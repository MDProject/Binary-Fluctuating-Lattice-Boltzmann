Given that we are approximating the stochastic Cahn-Hilliard Navier Stokes equation, we will compare our implementation to a binary FHD implementation in AMReX based on a staggered grid finite difference scheme \cite{Barker2023, bell_spherical_2025}. Bell et al. analyzed the thermal fluctuations of free-standing and sessile droplets where they simulated two fluid phases with physical properties equivalent to water \cite{bell_spherical_2025}. We base our unit system on the fluid properties and droplet size used in Bell et al. This nondimensionalization yields a lattice length scale of $\Delta x = 1.613 \times 10^{-9} \mathrm{m}$, a timescale of $\Delta t = 0.250 \times 10^{-12},\mathrm{s}$ and a mass scale of $\Delta m = 4.196 \times 10^{-24} \mathrm{kg}$. The correspondence between our physical and lattice-Boltzmann unit systems is summarized in Table \ref{tab:LB_unit_conversion}.

| Property | Real system | LB system |
|----------|-------------|-----------|
| Droplet radius  | $10 \times 10^{-9}\,\mathrm{m}$    | $6.2\,\Delta x$ |
| Fluid density   | $1 \times 10^{3}\,\mathrm{kg/m^3}$ | $1.0\,\Delta m / \Delta x^{3}$ |
| Fluid viscosity | $1 \times 10^{-2}\,\mathrm{Pa\,s}$ | $0.096\,\Delta m / (\Delta x\,\Delta t)$ |

From Table \ref{tab:Bell_vs_FLBM}, our Ohnesorge number, defined as $Oh = \frac{\eta}{\sqrt{\rho R \sigma}}$, where $\sigma$ is the surface tension, is $Oh = 1.231$. This differs from the value reported by Bell et al. $(Oh \approx 2)$ \cite{bell_spherical_2025}. The discrepancy arises primarily from the higher surface tension used in our model compared to that in Bell et al. Based on our unit conversions, the corresponding system temperature in our simulations is 126.76 K, which is higher than that employed by Bell et al. After correcting for the periodicity calculated as $P = 1 - 2.84\frac{R}{L}$, Bell et al.'s value is $P_{Bell} = 0.778$ while our value is $P_{FLBM} = 0.450$. Correcting for the difference in periodicity between Bell et al.'s and our simulation yields a diffusion value $\approx 11.3 \%$ greater than Bell et al.'s value, which also accounts for the difference in the simulation temperature between Bell et al.'s and our simulations.
