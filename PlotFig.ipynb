```python
#   IMG*** generation for each step
'''
/Users/dengxiaohui/Documents/AMReX/MultiComp/Lattice-Boltzmann/data_droplet/IMG_data_shs
han_alpha0_4.00_xi_0.0e+00_size32-32-32
'''
#######################################################################################################
########################################## Droplet ####################################################
#######################################################################################################
import yt
import os 

#file_path = "./lbm_data_shshan_alpha0_4_xi_mass1e-5_fittingEQ/";   # 1e-5mass is the kBT value
alpha0 = 4.
temperature = 0.
nx = 32
ny = 32
nz = 32
radius = 0.2

file_path = "./data_droplet_" + "{:.2f}".format(radius) + "/lbm_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) + "_size{:d}-{:d}-{:d}/".format(nx, ny, nz)
ts = yt.load(file_path + "plt*")
ac = 0

IMG_folder = "./data_droplet_" + "{:.2f}".format(radius) + "/IMG_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) \
            + "_size{:d}-{:d}-{:d}/".format(nx, ny, nz)
if(not os.path.isdir(IMG_folder)):
    os.mkdir(IMG_folder)

for fr in ts:
    slc = yt.SlicePlot(fr, "z", ('boxlib','rho')) #center=[0.5, 0.5, 0.5]
    slc.set_log(('boxlib','rho'), False)
    slc.set_zlim('rho', -0.5, 1.2)
    #slc.show()
    filename = IMG_folder + "fr" + str(ac).zfill(7) + ".png"
    slc.save(filename)
    ac = ac + 1
```
