```python
#   Image Viewer
import yt

pltfiledir = "./data_mixture_"
suffix = "hydrovars"
pltmfab = "equilibrium_rho_alpha0_0.00_size32-32-32"

file_path = pltfiledir + suffix + "/" + pltmfab
fr = yt.load(file_path)
print(fr.field_list)
slc = yt.SlicePlot(fr, "z", ('boxlib','rho_eq')) #center=[0.5, 0.5, 0.5]
slc.set_log(('boxlib','rho_eq'), False)
slc.set_zlim('rho_eq', 0.8, 1.2)
slc.show()
```

```python
#   IMG*** generation for each step
#######################################################################################################
########################################## Flate Interface ############################################
#######################################################################################################
import yt
import os 

alpha0 = 1.5
temperature = 0.
nx = 8
ny = 256
nz = 64

file_path = "./data_interface/lbm_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) + "_size{:d}-{:d}-{:d}/".format(nx, ny, nz)
ts = yt.load(file_path + "plt*")
ac = 0

IMG_folder = "./data_interface/IMG_data_shshan_alpha0_" + "{:.2f}".format(alpha0) +\
          "_xi_" + "{:.1e}".format(temperature) \
            + "_size{:d}-{:d}-{:d}/".format(nx, ny, nz)
if(not os.path.isdir(IMG_folder)):
    os.mkdir(IMG_folder)

for fr in ts:
    slc = yt.SlicePlot(fr, "x", ('boxlib','rho')) #center=[0.5, 0.5, 0.5]
    slc.set_log(('boxlib','rho'), False)
    slc.set_zlim('rho', -0.2, 2.5)
    #slc.show()
    filename = IMG_folder + "fr" + str(ac).zfill(7) + ".png"
    slc.save(filename)
    ac = ac + 1
```
