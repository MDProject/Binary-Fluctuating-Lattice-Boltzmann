### Pressure Tensor Formula -- Shan-Chen Model
####	Single Phase Case

$$\bm{F}\left(\bm{x}\right)=-G\psi\left(\bm{x}\right)\sum_{i}w_{i}\psi\left(\bm{x}+\bm{e}_{i}\right)\bm{e}_{i}$$

\begin{align*}
\psi\left(\bm{x}+\bm{e}_{i}\right)	&=\psi\left(\bm{x}\right)+\bm{e}_{i}\cdot\nabla\psi\left(\bm{x}\right)+\frac{1}{2}\left(\bm{e}_{i}\cdot\nabla\right)^{2}\psi\left(\bm{x}\right)+\frac{1}{3!}\left(\bm{e}_{i}\cdot\nabla\right)^{3}\psi\left(\bm{x}\right)+\cdot\cdot\cdot \\
	&=\psi\left(\bm{x}\right)+\partial_{\alpha}\psi\left(\bm{x}\right)e_{i\alpha}+\frac{1}{2}\partial_{\alpha}\partial_{\beta}\psi\left(\bm{x}\right)e_{i\alpha}e_{i\beta}+\frac{1}{3!}\partial_{\alpha}\partial_{\beta}\partial_{\gamma}\psi\left(\bm{x}\right)e_{i\alpha}e_{i\beta}e_{i\gamma}+\cdot\cdot\cdot
\end{align*}

\begin{align*}
F_{\alpha}\left(\bm{x}\right)	&=-G\psi\left(\bm{x}\right)\sum_{i}w_{i}\psi\left(\bm{x}+\bm{e}_{i}\right)e_{i\alpha} \\
	&=-G\psi\left(\bm{x}\right)\sum_{i}w_{i}\left[\psi\left(\bm{x}\right)+\partial_{\beta}\psi\left(\bm{x}\right)e_{i\beta}+\frac{1}{2}\partial_{\beta}\partial_{\gamma}\psi\left(\bm{x}\right)e_{i\beta}e_{i\gamma}+\frac{1}{3!}\partial_{\beta}\partial_{\gamma}\partial_{\delta}\psi\left(\bm{x}\right)e_{i\beta}e_{i\gamma}e_{i\delta}+\cdot\cdot\cdot\right]e_{i\alpha} \\
	&\approx-G\psi^{2}\left(\bm{x}\right)\sum_{i}w_{i}e_{i\alpha}-G\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\sum_{i}w_{i}e_{i\alpha}e_{i\beta}-\frac{1}{2}G\psi\left(\bm{x}\right)\partial_{\beta}\partial_{\gamma}\psi\left(\bm{x}\right)\sum_{i}w_{i}e_{i\alpha}e_{i\beta}e_{i\gamma}-\frac{1}{6}G\psi\left(\bm{x}\right)\partial_{\beta}\partial_{\gamma}\partial_{\delta}\psi\left(\bm{x}\right)\sum_{i}w_{i}e_{i\alpha}e_{i\beta}e_{i\gamma}e_{i\delta} \\
	&=0-Gc_{s}^{2}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\delta_{\alpha\beta}-0-\frac{1}{6}Gc_{s}^{4}\psi\left(\bm{x}\right)\partial_{\beta}\partial_{\gamma}\partial_{\delta}\psi\left(\bm{x}\right)\left(\delta_{\alpha\beta}\delta_{\gamma\delta}+\delta_{\alpha\gamma}\delta_{\beta\delta}+\delta_{\alpha\delta}\delta_{\beta\gamma}\right) \\
	&=-Gc_{s}^{2}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\delta_{\alpha\beta}-\frac{1}{2}Gc_{s}^{4}\psi\left(\bm{x}\right)\partial_{\beta}\partial_{\gamma}\partial_{\gamma}\psi\left(\bm{x}\right)\delta_{\alpha\beta}
\end{align*}


$$\bm{F}\left(\bm{x}\right)=-\nabla\cdot\bm{P}=-\partial_{\beta}P_{\alpha\beta}$$

\begin{align*}
\partial_{\beta}P_{\alpha\beta}	&=Gc_{s}^{2}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\delta_{\alpha\beta}+\frac{1}{2}Gc_{s}^{4}\psi\left(\bm{x}\right)\partial_{\beta}\partial_{\gamma}\partial_{\gamma}\psi\left(\bm{x}\right)\delta_{\alpha\beta} \\
	&=\frac{1}{2}Gc_{s}^{2}\partial_{\beta}\psi^{2}\left(\bm{x}\right)\delta_{\alpha\beta}+\frac{1}{2}Gc_{s}^{4}\partial_{\beta}\left[\psi\left(\bm{x}\right)\partial_{\gamma}\partial_{\gamma}\psi\left(\bm{x}\right)\delta_{\alpha\beta}\right]-\frac{1}{2}Gc_{s}^{4}\partial_{\beta}\psi\left(\bm{x}\right)\partial_{\gamma}\partial_{\gamma}\psi\left(\bm{x}\right)\delta_{\alpha\beta} \\
	&=\frac{1}{2}Gc_{s}^{2}\partial_{\beta}\psi^{2}\left(\bm{x}\right)\delta_{\alpha\beta}+\frac{1}{2}Gc_{s}^{4}\partial_{\beta}\left[\psi\left(\bm{x}\right)\nabla^{2}\psi\left(\bm{x}\right)\delta_{\alpha\beta}\right]-\frac{1}{2}Gc_{s}^{4}\left\{ \partial_{\beta}\left[\partial_{\alpha}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\right]-\partial_{\alpha}\partial_{\beta}\psi\left(\bm{x}\right)\cdot\partial_{\beta}\psi\left(\bm{x}\right)\right\} \\
	&=\frac{1}{2}Gc_{s}^{2}\partial_{\beta}\psi^{2}\left(\bm{x}\right)\delta_{\alpha\beta}+\frac{1}{2}Gc_{s}^{4}\partial_{\beta}\left[\psi\left(\bm{x}\right)\nabla^{2}\psi\left(\bm{x}\right)\delta_{\alpha\beta}\right]-\frac{1}{2}Gc_{s}^{4}\partial_{\beta}\left[\partial_{\alpha}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\right]+\frac{1}{4}Gc_{s}^{4}\partial_{\alpha}\left|\nabla\psi\left(\bm{x}\right)\right|^{2}
\end{align*}

\begin{align*}
P_{\alpha\beta}	&=\frac{1}{2}Gc_{s}^{2}\psi^{2}\left(\bm{x}\right)\delta_{\alpha\beta}+\frac{1}{2}Gc_{s}^{4}\left[\psi\left(\bm{x}\right)\nabla^{2}\psi\left(\bm{x}\right)\delta_{\alpha\beta}\right]-\frac{1}{2}Gc_{s}^{4}\left[\partial_{\alpha}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)\right]+\frac{1}{4}Gc_{s}^{4}\left|\nabla\psi\left(\bm{x}\right)\right|^{2}\delta_{\alpha\beta} \\
	&=\left[\frac{1}{2}Gc_{s}^{2}\psi^{2}\left(\bm{x}\right)+\frac{1}{2}Gc_{s}^{4}\psi\left(\bm{x}\right)\nabla^{2}\psi\left(\bm{x}\right)+\frac{1}{4}Gc_{s}^{4}\left|\nabla\psi\left(\bm{x}\right)\right|^{2}\right]\delta_{\alpha\beta}-\frac{1}{2}Gc_{s}^{4}\partial_{\alpha}\psi\left(\bm{x}\right)\partial_{\beta}\psi\left(\bm{x}\right)
\end{align*}

```python
from scipy.optimize import curve_fit

step0 = 0
step1 = 20000
plot_int = 100
temperature = 0.
nx, ny, nz = 32, 32, 32
alpha0 = 1.7 #2. #1.5
radius = 0.28
rho_hi = 3.0 #3.0

def get_center_of_mass(rho_field, L, cell_centered=True):
    """
    Compute center of mass for a 3D density field using trapezoidal integration
    ----------
    rho_field : ndarray
        3D density field (shape: nx, ny, nz).
    L : tuple of float
        Physical dimensions of the domain (Lx, Ly, Lz).
    cell_centered : bool
        If True, grid is cell based; else node based.
    Returns
    -------
    vec_com : numpy.ndarray
        Center of mass [x_cm, y_cm, z_cm].
    """
    nx, ny, nz = rho_field.shape
    if cell_centered:
        dx, dy, dz = L[0] / nx, L[1] / ny, L[2] / nz
        x = (np.arange(nx) + 0.5) * dx
        y = (np.arange(ny) + 0.5) * dy
        z = (np.arange(nz) + 0.5) * dz
    else:  # node-centered
        dx, dy, dz = L[0] / (nx - 1), L[1] / (ny - 1), L[2] / (nz - 1)
        x = np.arange(nx) * dx
        y = np.arange(ny) * dy
        z = np.arange(nz) * dz
    X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

    # Total mass (weighted integral of density)
    mass = np.sum(rho_field * dx * dy * dz)
    # Weighted coordinate integrals
    x_cm = np.sum(X * rho_field * dx * dy * dz) / mass
    y_cm = np.sum(Y * rho_field * dx * dy * dz) / mass
    z_cm = np.sum(Z * rho_field * dx * dy * dz) / mass

    return np.array([x_cm, y_cm, z_cm])


def func_density_dist(rho_field, L, r0=None):
    """
    Flatten the 3D density field into a 1D array and compute the relative distance from the center.
    ----------
    rho_field : np.ndarray
        The 3D density field (shape: [Nz, Ny, Nx]).
    L : tuple of float
        Physical dimensions of the domain (Lx, Ly, Lz), used to convert from dimensionless to physical units.
    Returns
    -------
    np.ndarray
        Flattened 1D array of the density field & relative distance from the center.
    """
    if r0 is None:
        r0 = get_center_of_mass(rho_field, L)
    Nz, Ny, Nx = rho_field.shape
    # Create 3D grid of coordinates
    dx, dy, dz = L[0] / Nx, L[1] / Ny, L[2] / Nz
    x = (np.arange(Nx) + 0.5) * dx
    y = (np.arange(Ny) + 0.5) * dy
    z = (np.arange(Nz) + 0.5) * dz
    r0 *= dx.units  # Convert from dimensionless to physical units !
    X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
    r = np.sqrt(((X - r0[0])**2) + ((Y - r0[1])**2) + ((Z - r0[2])**2))
    return rho_field.flatten(), r.flatten()

def read_frame_densities(file_path):
    ts = yt.load(file_path)
    profile_data = ts.covering_grid(
        level=0, left_edge=ts.domain_left_edge, dims=ts.domain_dimensions
    )
    profile1 = profile_data['boxlib', 'rho']
    profile2 = profile_data['boxlib', 'phi']
    # grid lattice length
    box  = ts.domain_dimensions
    L = ts.domain_right_edge - ts.domain_left_edge
    dr = L/box
    return profile1, profile2, L, box, dr

def fit_droplet_radius(rho_field, L, r0):
    """
    Fit droplet radius, width & max/min density by using a tanh function. See above.
    Returns: (rho_hi_fit, rho_lo_fit, radius_fit, W_fit) tuples
    """
    # Define the fitting function
    def tanh_profile(r, rho_hi, rho_lo, radius, W):
        return rho_hi - (rho_hi - rho_lo)/2 * (1 + np.tanh((r - radius)/W))

    rho_field_flatten, r = func_density_dist(rho_field, L, r0)
    # Fit the density profile to the tanh function
    popt, _ = curve_fit(tanh_profile, r, rho_field_flatten, p0=[rho_field_flatten.max(), rho_field_flatten.min(), 0.5, 0.5])
    rho_hi_fit, rho_lo_fit, radius_fit, W_fit = popt
    return rho_hi_fit, rho_lo_fit, radius_fit, W_fit

radius_list = []
rho_hi_list = []
rho_lo_list = []
for step in range(step0, step1+1, plot_int):
    file_path = plot_file_root + "/data_droplet_density_{:.2f}_alpha0_{:.2f}_r{:.2f}_size{:d}-{:d}-{:d}/lbm_data_shshan_alpha0_"\
        .format(rho_hi, alpha0, radius, nx, ny, nz) +\
      "{:.2f}".format(alpha0) + "_xi_" + "{:.1e}".format(temperature) + "_size{:d}-{:d}-{:d}/plt".format(nx,ny,nz) + f"{step:07d}"
    rho_field, phi_field, L, box, dr = read_frame_densities(file_path)
    r0 = get_center_of_mass(rho_field, L)
    rho_hi_fit, rho_lo_fit, radius_fit, W_fit = fit_droplet_radius(rho_field, L, r0)
    radius_list.append(radius_fit)
    rho_hi_list.append(rho_hi_fit)
    rho_lo_list.append(rho_lo_fit)
    #print("step = {:d}, fitted radius = {:.4f}, W = {:.4f}, rho_hi = {:.4f}, rho_lo = {:.4f}".format(step, radius_fit/dr[0], W_fit/dr[0], rho_hi_fit, rho_lo_fit))

radius_array = np.array(radius_list)
N  = radius_array.size
step = np.linspace(0, N-1, N)

# Plotting the data
plt.figure(figsize=(8, 6), dpi=100)  # Optional: Set the figure size
#fig, ax = plt.subplots()
plt.plot(step, radius_array, color='b')  # Create a line plot
plt.yscale('linear')
# Add labels and title
plt.xlabel("steps")  # X-axis label
plt.ylabel("radius")  # Y-axis label
plt.title("radius ~ steps, alpha0 = {:.2f}, size = [{:d},{:d},{:d}]".format(alpha0, nx, ny, nz))  # Plot title

# Show the plot
plt.show()

print(radius_array[-1])
print(rho_hi_list[-1])
print(rho_lo_list[-1])
```

### Pressure difference

The pressure difference caused by SC force is
$$\bm{F}^{SC}\left(\bm{x}\right):=\bm{F}^{\sigma}\left(\bm{x}\right)+\bm{F}^{\sigma^{\prime}}\left(\bm{x}\right)=-\nabla\cdot\bm{P}^{SC}$$
$$-\frac{\partial}{\partial x}P^{SC}=F^{SC}_x,\ \ \ P\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)-P\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)=\int_0^{\frac{L_x}{2}} F^{SC}_xdx$$

In the following code, **deltaP[*]_SC** denotes the pressure difference calculated by integral along $x$ axis (RHS); **Pressure_SC** denotes the pressure difference calculated by $G_{\sigma\sigma^{\prime}}c_{s}^{2}\psi^{\sigma}\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)\psi^{\sigma^{\prime}}\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)-G_{\sigma\sigma^{\prime}}c_{s}^{2}\psi^{\sigma}\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)\psi^{\sigma^{\prime}}\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)$ (LHS).

**DeltaP_array** denotes the **total pressure** difference (**ideal gas**+**SC**) array calculated by
$$\rho_{\text{total}}\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)c_s^2 - \rho_{\text{total}}\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)c^2_s + G_{\sigma\sigma^{\prime}}c_{s}^{2}\psi^{\sigma}\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)\psi^{\sigma^{\prime}}\left(\frac{L_x}{2},\frac{L_y}{2},\frac{L_z}{2}\right)-G_{\sigma\sigma^{\prime}}c_{s}^{2}\psi^{\sigma}\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)\psi^{\sigma^{\prime}}\left(0,\frac{L_y}{2},\frac{L_z}{2}\right)$$
